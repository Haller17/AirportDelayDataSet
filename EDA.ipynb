```python
#Changing the data in the colums from Minutes to Hours

import pandas as pd

df = pd.read_csv("Airline_Delay_Cause.csv")

delay_cols = [
    'arr_delay',
    'carrier_delay',
    'weather_delay',
    'nas_delay',
    'security_delay',
    'late_aircraft_delay'
]

df[delay_cols] = df[delay_cols] / 60

df[delay_cols] = df[delay_cols].round(2)

df.to_csv("Airline_Delay_Cause_hours.csv", index=False)
```

### 5. Univariate Analysis

```python
# Define features for visualization (Choosing the numerical features)
num_features = ['carrier_ct',
    'weather_ct',
    'nas_ct',
    'security_ct',
    'late_aircraft_ct',
    'arr_delay']

# Create a 2x3 grid for visualization
fig, ax = plt.subplots(2, 3, figsize=(18, 12))

# Iterate over features and plot
for i, feature in enumerate(num_features):
    row, col = divmod(i, 3)
    sns.histplot(df[feature], kde=True, bins=30, ax=ax[row, col])
    ax[row, col].set_title(f'Distribution of {feature}')
    
plt.tight_layout()
plt.show()
```

```python
# Create boxplots for numerical variables
fig, ax = plt.subplots(3, 2, figsize=(18, 15))

for i, feature in enumerate(num_features):
    row, col = divmod(i, 2)
    sns.boxplot(y=df[feature], ax=ax[row, col])
    ax[row, col].set_title(f'Boxplot of {feature}')

plt.tight_layout()
plt.show()
```

```python
df.columns
```

```python
# Create violin plots for numerical variables
fig, ax = plt.subplots(3, 2, figsize=(18, 15))

for i, feature in enumerate(num_features):
    row, col = divmod(i, 2)
    sns.violinplot(y=df[feature], ax=ax[row, col])
    ax[row, col].set_title(f'Violin Plot of {feature}')

plt.tight_layout()
plt.show()
```

```python
# Define categorical features for visualization
cat_features = ['carrier_name',
    'airport',
    'month',
    'year']

# Create a 2x3 grid for visualization
fig, ax = plt.subplots(2, 3, figsize=(18, 12))

# Iterate over categorical features and plot
for i, feature in enumerate(cat_features):
    row, col = divmod(i, 3)
    sns.countplot(data=df, x=feature, order=df[feature].value_counts().index, ax=ax[row, col])
    ax[row, col].set_title(f'Distribution of {feature}')
    ax[row, col].tick_params(axis='x', rotation=40)  # Rotate x-axis labels for better readability
    
plt.tight_layout()
plt.show()
```

### 6. Bivariate Analysis

```python
# Bar Chart for numerical vs. categorical
top_airports = df.groupby('airport')['arr_delay'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
top_airports.plot(kind='bar')
plt.title("Top 10 Airports With Highest Total Delay Hours")
plt.ylabel("Total Delay Hours")
plt.xticks(rotation=45)
plt.show()

# Pie Chart for categorical vs. numerical
delay_causes = df[['carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']].sum()

plt.figure(figsize=(8,8))
plt.pie(delay_causes, labels=delay_causes.index, autopct='%1.1f%%')
plt.title("Proportion of Delay Causes (Hours)")
plt.show()

# Line Chart for numerical vs. categorical
monthly_delay = df.groupby('month')['arr_delay'].mean()

plt.figure(figsize=(10,6))
monthly_delay.plot(marker='o')
plt.title("Average Delay Hours by Month")
plt.xlabel("Month")
plt.ylabel("Average Delay Hours")
plt.show()

# Horizontal bar chart for delay causes
delay_types = df[['carrier_delay','weather_delay','nas_delay','security_delay','late_aircraft_delay']].sum()

plt.figure(figsize=(10,6))
delay_types.sort_values().plot(kind='barh')

plt.title("Total Delay Hours by Cause")
plt.xlabel("Delay Hours")
plt.ylabel("Delay Type")
plt.show()

# Scatter plot for numerical vs numerical
plt.figure(figsize=(8,6))

plt.scatter(df['arr_flights'], df['arr_delay'], alpha=0.3)

plt.xlabel("Number of Flights")
plt.ylabel("Total Delay Hours")
plt.title("Flights vs Total Delay Hours")

plt.show()
```
